```python
# unpack string
def unpack_passwords(passwords):
    unpacked_string = []
    for i in passwords:
        separated_key_1 = i.split('-')
        separated_key_2 = separated_key_1[1].split(' ')
        x = [int(separated_key_1[0]), int(separated_key_2[0]), separated_key_2[1][0], separated_key_2[2]]
        unpacked_string.append(x)
    return unpacked_string
```
