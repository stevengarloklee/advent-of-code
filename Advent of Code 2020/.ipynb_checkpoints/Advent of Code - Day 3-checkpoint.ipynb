```python
class tree_finder:
    def __init__(self, initial_map):
        self.initial_map = initial_map
        
    def widen_map(self, right_moves):
        wider_map = self.initial_map
        required_width = len(self.initial_map) * right_moves
        
        while len(wider_map[0]) < required_width:
            wider_map = [x + x for x in wider_map]
        
        return wider_map

    def mark_x_o(self, wider_map, right_moves):
        new_map = []
        for i,x in enumerate(wider_map):
            tree_position = i * right_moves
            characters = list(x)      

            if (characters[tree_position] == '.' and tree_position < len(characters)):
                characters[tree_position] = 'O'
            elif (characters[tree_position] == '#' and tree_position < len(characters)):
                    characters[tree_position] = 'X'
            else:
                pass
            new_row = ''.join(characters)
            new_map.append(new_row)
        return new_map
    
    def mark_x_o_right_down(self, wider_map, right_moves, down_moves):
        new_map = []
        for i,x in enumerate(wider_map):
            tree_position = math.floor((i+1) / down_moves)
            characters = list(x)      

            if (characters[tree_position] == '.' and tree_position < len(characters) and (i+1)%down_moves != 0 and i != 0 and i !=1):
                characters[tree_position] = 'O'
            elif (characters[tree_position] == '#' and tree_position < len(characters) and (i+1)%down_moves != 0 and i != 0 and i !=1):
                    characters[tree_position] = 'X'
            else:
                pass
            new_row = ''.join(characters)
            new_map.append(new_row)
        return new_map
                
    def count_x(self, new_map):
        x_count = 0
        for i in new_map:
            characters = list(i)
            for x in characters:
                if x == 'X':
                    x_count += 1
                else:
                    pass
        return x_count
```

```python
wider_map = tf.widen_map(3)
```

```python
# down 1
# new_map = tf.mark_x_o(wider_map,3)
# down 2
new_map = tf.mark_x_o_right_down(wider_map, 1, 2)
```

--- Part Two ---
Time to check the rest of the slopes - you need to minimize the probability of a sudden arboreal stop, after all.

Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:

Right 1, down 1.
Right 3, down 1. (This is the slope you already checked.)
Right 5, down 1.
Right 7, down 1.
Right 1, down 2.
In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.

What do you get if you multiply together the number of trees encountered on each of the listed slopes?

```python
right_moves_list = [1,3,5,7]
```

```python
number_of_trees = []
for i in right_moves_list:
    tf = tree_finder(real_map)
    wider_map = tf.widen_map(i)
    new_map = tf.mark_x_o(wider_map,i)
    number_of_trees.append(tf.count_x(new_map))
```

```python
tf = tree_finder(real_map)
wider_map = tf.widen_map(1)
new_map = tf.mark_x_o_right_down(wider_map,1,2)
number_of_trees.append(tf.count_x(new_map))
```

```python
print (number_of_trees)
```

```python
result = 1
for x in number_of_trees:
    result = result * x
print (result)
```
